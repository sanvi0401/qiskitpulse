```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
```

```python
bell = QuantumCircuit(2)
bell.h(0)
bell.cx(0, 1)
bell.x(1)
bell.z(0)     

print("Bell state |Ψ->")
display(bell.draw("mpl"))
```

```python
qr = QuantumRegister(3, 'q')
cr = ClassicalRegister(2, 'c')
tele = QuantumCircuit(qr, cr)

# Input state
tele.h(qr[0])
tele.t(qr[0])

# Bell |Ψ->
tele.h(qr[1])
tele.cx(qr[1], qr[2])
tele.x(qr[2])
tele.z(qr[1])

# Bell measurement
tele.cx(qr[0], qr[1])
tele.h(qr[0])
tele.measure(qr[0], cr[0])
tele.measure(qr[1], cr[1])

# Corrections
tele.x(qr[2]).c_if(cr, 1)
tele.z(qr[2]).c_if(cr, 2)
tele.x(qr[2]).c_if(cr, 3)
tele.z(qr[2]).c_if(cr, 3)

print("Teleportation using |Ψ->")
display(tele.draw("mpl"))
```
