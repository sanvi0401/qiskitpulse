```python
%matplotlib inline
from qiskit import pulse
from qiskit.pulse import DriveChannel, AcquireChannel, MemorySlot, Gaussian
from qiskit.visualization.pulse_v2 import draw as pulse_draw
import matplotlib.pyplot as plt
```

```python
duration=64
sigma=16
p_h=Gaussian(duration,0.3,sigma)
p_cx=Gaussian(duration,0.35,sigma)
d0=DriveChannel(0)
d1=DriveChannel(1)

with pulse.build(name="phi_plus_bell") as bell_pulse:
    pulse.play(p_h,d0)
    pulse.delay(16,d0)
    pulse.play(p_cx,d0)
    pulse.play(p_cx,d1)

bell_pulse
```

```python
fig=pulse_draw(bell_pulse)
fig.set_size_inches(14,4)
plt.show()
```

```python
d0=DriveChannel(0)
d1=DriveChannel(1)
d2=DriveChannel(2)
a0=AcquireChannel(0)
a1=AcquireChannel(1)
a2=AcquireChannel(2)
m0=MemorySlot(0)
m1=MemorySlot(1)
m2=MemorySlot(2)

with pulse.build(name="phi_plus_teleport") as tele_pulse:
    pulse.play(p_h,d0)
    pulse.delay(16,d0)
    pulse.play(p_cx,d0)
    pulse.play(p_cx,d1)
    pulse.play(Gaussian(duration,0.3,sigma),d1)
    pulse.play(Gaussian(duration,0.3,sigma),d2)
    pulse.acquire(128,a0,m0)
    pulse.acquire(128,a1,m1)
    pulse.play(Gaussian(duration,0.3,sigma),d2)
    pulse.play(Gaussian(duration,0.3,sigma),d2)
    pulse.acquire(128,a2,m2)

tele_pulse
```

```python
fig=pulse_draw(tele_pulse)
fig.set_size_inches(16,6)
plt.show()
```
